```python
MONGO_URI = 'mongodb://localhost:27017/'

mongo_client = MongoClient(MONGO_URI)
kdd_db = mongo_client.kdd2013
author_collection = kdd_db.author
```

```python
author_cache = list(author_collection.find())
```

```python
affiliations_data = [
    list(map(prepare_affiliation, author['affiliations']))
    for author
    in author_cache
]
```

```python
import pprint

pprint.pprint(affiliations_data[:2])
```

```python
import itertools
from textwrap import dedent

for affiliations_of_one_author in affiliations_data[:2]:
    print('=' * 20)
    for (a, b) in itertools.combinations(affiliations_of_one_author, 2):
        l = levenshtein(a, b)
        ls = levenshtein_sort(a, b)
        if l < ls:
            print(dedent(f"""
                {'-' * 10}
                a: {a}
                b: {b}
                l: {l}
                ls: {ls}
            """).strip())
#         if l < ls:
#             print(a, b, l, ls)
```
